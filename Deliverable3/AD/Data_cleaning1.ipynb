```python
# Change the Genders from str to int by gving them values of 0,1,2
def fix_gender_column():
    merged_data["What is your gender?"] = merged_data["What is your gender?"].str.lower()
    merged_data["What is your gender?"].replace(['female'], '0', inplace=True)
    merged_data["What is your gender?"].replace(["male"], "1", inplace=True)
    merged_data["What is your gender?"].replace(["gender non-binary/other"], "2", inplace=True)
    
fix_gender_column()
#Should have three categories
merged_data["What is your gender?"].unique()
```

```python
# Make the "Have you ever been diagnosed with a mental health disorder?" Column have 0 and 1 value. Anything NaN will be treated as zero
def fix_disorder_column():
    merged_data["Have you ever been diagnosed with a mental health disorder?"] = merged_data["Have you ever been diagnosed with a mental health disorder?"].str.lower()
    merged_data["Have you ever been diagnosed with a mental health disorder?"].replace(['No','no'], '0', inplace=True)
    merged_data["Have you ever been diagnosed with a mental health disorder?"].replace(["Yes", "yes"], "1", inplace=True)
    merged_data["Have you ever been diagnosed with a mental health disorder?"].fillna(value="0", inplace=True)

    
fix_disorder_column()
#Should have three categories
merged_data["Have you ever been diagnosed with a mental health disorder?"].unique()
```

```python
# Make the "Have you had a mental health disorder in the past?" have a 0 and 1 value also.
#Those in the unsure will be consider a no

def fix_past_column():
    merged_data["Have you had a mental health disorder in the past?"] = merged_data["Have you ever been diagnosed with a mental health disorder?"].str.lower()
    merged_data["Have you had a mental health disorder in the past?"].replace(['No','no',"I don't know"], '0', inplace=True)
    merged_data["Have you had a mental health disorder in the past?"].replace(["Yes", "Possibly"], "1", inplace=True)
    merged_data["Have you had a mental health disorder in the past?"].fillna(value="0", inplace=True)

    
fix_past_column()
#Should have three categories
merged_data["Have you ever been diagnosed with a mental health disorder?"].unique()
```

```python
# Make the "Do you have a family history of mental illness?" Columns will have  0 and 1 value ,for those with 
#"I don't know will be treated as no"


def fix_past_column():
    merged_data["Do you have a family history of mental illness?"] = merged_data["Do you have a family history of mental illness?"].str.lower()
    merged_data["Do you have a family history of mental illness?"].replace(['No','no',"I don't know","i don't know"], '0', inplace=True)
    merged_data["Do you have a family history of mental illness?"].replace(["Yes", "Possibly",'yes'], "1", inplace=True)
    merged_data["Do you have a family history of mental illness?"].fillna(value="0", inplace=True)

    
fix_past_column()
#Should have three categories
merged_data["Do you have a family history of mental illness?"].unique()
```
