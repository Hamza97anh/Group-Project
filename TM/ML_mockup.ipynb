```python
# Logistic Regression - this model will predict whether or not an individual has a mental health disorder (output: yes or no)
```

```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
import numpy as np
```

```python
# Load data
csv = pd.read_csv('./SB/mental-heath-in-tech-2016_20161114.csv')
data = pd.DataFrame(csv)
data.head(50)
```

```python
# Rename ind and dep variable columns
renamed_data = data.rename(columns = {'Is your employer primarily a tech company/organization?':'techOrg',\
                                      'Has your employer ever formally discussed mental health (for example, as part of a wellness campaign or other official communication)?':'formalDiscussion',\
                                      'Have you had a mental health disorder in the past?':'mentalHealthDisorder',\
                                      'What is your age?':'age'})
renamed_data.head()
```

```python
# # Convert col to strings
# renamed_data.formalDiscussion = renamed_data.formalDiscussion.astype(str)
```

```python
# # Clean to only 'Yes' or 'No'
# new_col = []

# # Loop through dataframe
# for data in renamed_data.values:
    
#     if (str(data[6]).lower() != 'yes') and (str(data[6]).lower() != 'no'):
#         data[6] = None
#     new_col.append(data[6])

# # Create new df
# renamed_data.formalDiscussion = new_col

        
df = renamed_data.dropna(subset=['techOrg'])
renamed_data.head(10)
```

```python
## Potential independent variables:
# Age
# Gender
# Self Employment
# Remote Working Situation

## Dependent vaibale:
# Whether they have been diagnosed with a mental health disorder by a medical professional
```

```python
# Potential independent variable = whether manager has had formal discussion about mental health importance and 
# opportunities with employee
X = df.techOrg.values.reshape(-1,1)
```

```python
# Dependent vairable = whether the employee has been diagnosed with a mental health disorder
y = df.mentalHealthDisorder.values.reshape(-1,1)
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
```

```python
# Create logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
```

```python
## Train data
classifier.fit(X_train, y_train.ravel())
```

```python
## Predict outcomes for test data
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test.ravel()})
```

```python
# Validate model with test data
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```
